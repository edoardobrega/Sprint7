```python
"""IMC"""


#Función para el cálculo del IMC y su categoría
#Pasándole los parámetros de peso y altura ya validados en otra función, esta función realiza el cálculo del IMC y determina su categoría.
#La función siempre devuelve estas dos variables.

def calc_imc(peso, altura):
    imc = peso / (altura ** 2)
    if imc < 18.5: cat = "Bajo Peso"
    elif 18.5 <= imc < 25: cat = "Peso Normal"
    elif 25 <= imc < 30: cat = "Sobrepeso"
    else: cat = "Obesidad"
    return imc, cat

#Función para la introducción del peso 
#Esta función solicita al usuario que ingrese el peso y verifica que esté dentro del rango aceptado.  
#Si no está dentro del rango aceptado, informa al usuario y vuelve a solicitar el peso.  
#La función devuelve el peso aceptable o no devuelve nada si el usuario presiona la tecla Esc.

def obtener_peso():

    peso_valido = False

    while not peso_valido:

        peso = input("Ingresa el peso en 'Kg' o presiona ESC para volver al principio: ").replace(",", ".").lower().replace("kg", "")

        try:
            valor_peso = float(peso)
            
            if 29.9 <= valor_peso <= 250.1: peso_valido = True ; return valor_peso
            
            else: print(f"El valor ingresado para el peso es {peso}. La masa corporal se certifica solo para valores entre 30 kg y 250 kg.")

        except ValueError:
            
            if keyboard.is_pressed('esc'): print("Has presionado Esc. ") ; peso_valido = True ; return None

            elif not peso: print("No has ingresado el peso.")
            
            else: print(f"El valor ingresado no es valido: {peso}. Ingresa otra vez el peso.")
           
#Función para la introducción de la altura 
#Esta función solicita al usuario que ingrese la altura y verifica que esté dentro del rango aceptado.  
#Si no está dentro del rango aceptado, informa al usuario y vuelve a solicitar la altura.  
#La función devuelve la altura aceptable o no devuelve nada si el usuario presiona la tecla Esc.

def obtener_altura():

    altura_valida = False

    while not altura_valida:

        altura = input("Ingresa la altura en 'm' o presiona ESC para volver al principio: ").replace(",", ".").lower().replace("m", "")

        try:
            valor_altura = float(altura)

            if 0.9 < valor_altura < 2.6: altura_valida = True ; return valor_altura

            else:

                print(f"El valor ingresado para la altura es {valor_altura}. La masa corporal se certifica solo para valores entre 1 m y 2.5 m.")

        except ValueError:
            
            if keyboard.is_pressed('esc'): print("Has presionado Esc. ") ; altura_valida = True ; return None

            elif not altura: print("No has ingresado la altura.")

            else: print(f"El valor ingresado no es valido: {altura}.")

#Esta función llama a las funciones de introducción de peso y altura y las combina para ser utilizadas en el cálculo del IMC y de la categoría.
#La función devuelve ambos valores o no devuelve nada en caso de que el usuario haya presionado Esc durante la introducción de uno de los dos valores.

def obtener_datos_imc():

    peso = obtener_peso()

    if peso is not None:

        altura = obtener_altura()

        if altura is not None:

            return peso, altura
        
    return None, None

#Función principal para agrupar las demás funciones y mostrar el resultado en pantalla.

def main_imc():

    print("Programa para el calculo del Índice de masa corporal.\n")

    sigo_con_programa = 'y'

    while sigo_con_programa == 'y':

        respuesta = input("Quieres calcular el IMC de una persona? (y/n): ").strip().lower()
        
        if keyboard.is_pressed('esc') or respuesta == 'n': print("Salida del programa.") ; sigo_con_programa = 'n'
        
        elif respuesta == 'y':
            
            peso, altura = obtener_datos_imc()
            
            if peso is not None and altura is not None:
                
                imc, categoria = calc_imc(peso, altura)
                
                print(f"Datos de peso y altura recibidos: {peso}, {altura}\nIMC = {imc:.2f}\nCategoria = {categoria}.")

                sigo_con_programa = 'n'
        
        else:
            
            print("Respuesta no reconocida. Presiona 'y', 'n' o 'Esc'.")

import keyboard

main_imc()
```

```python
"""Temperaturas"""

#Función para la conversión de la temperatura ingresada
#Esta función convierte la temperatura ingresada que ya fue verificada en otra función.
#Verifica de qué unidad de medida se trata y devuelve las conversiones correspondientes.

def convertir_temperatura(temp_in):

    if temp_in[-1].upper() == 'C':

        celsius =  float(temp_in[:-1])
        
        return f"{celsius}° Celsius equivalen a : {(celsius * 9/5) + 32:.2f}° Fahrenheit y equivalen a : {celsius + 273.15:.2f}° Kelvin."
    
    elif temp_in[-1].upper() == 'F': 

        fahrenheit =  float(temp_in[:-1])

        return f"{fahrenheit}° Fahrenheit equivalen a : {(fahrenheit - 32) * 5/9:.2f}° Celsius y equivalen a : {(fahrenheit - 32) * 5 / 9 + 273.15:.2f}° Kelvin."
    
    elif temp_in[-1].upper() == 'K': 

        kelvin =  float(temp_in[:-1])

        return f"{kelvin}° Kelvin equivalen a : {kelvin - 273.15:.2f}° Celsius y equivalen a : {(kelvin - 273.15) * 9 / 5 + 32:.2f}° Fahrenheit."

#Función para la introducción y verificación de la temperatura
#La función solicita al usuario que ingrese una temperatura.
#En caso de que el usuario introduzca una temperatura inferior al cero absoluto, muestra un mensaje de error y vuelve a pedir al usuario que ingrese la temperatura.
#La función devuelve una temperatura válida o no devuelve nada si el usuario presiona la tecla Esc.

def obtener_temperatura():
    
    temperatura_valida = False 
    
    while not temperatura_valida:
        
        temp_in = input("Ingresa la temperatura que quieres convertir con su unidad de medida\n(ejemplos: 25C, 77F, 300K): ").strip().replace(',', '.').replace("°", "")
        
        try: temp_valor = float(temp_in[:-1])  

        except ValueError: 

            if keyboard.is_pressed('esc'): print("Has presionado Esc.") ; temperatura_valida = True ; return None

            elif not temp_in: print("No has ingresado la temperatura.") 
            
            print("Ingresa un valor numerico valido para la temperatura seguido de su unidad de medida.") ; continue

        if temp_in[-1].upper() not in ['C', 'F', 'K']: print("Unidad de medida no reconocida. Recuerda añadir C, F o K despues del valor de la temperatura.")

        else:

            if temp_in[-1].upper() == 'C' and -273.15 <= temp_valor : temperatura_valida = True ; return temp_in
            
            elif temp_in[-1].upper() == 'F' and -459.67 <= temp_valor : temperatura_valida = True ; return temp_in
            
            elif temp_in[-1].upper() == 'K' and 0 <= temp_valor : temperatura_valida = True ; return temp_in
            
            else:
                
                if temp_in[-1].upper() == 'C': print("Las temperaturas en Celsius tienen que ser mayor de -273.15°C")
                
                if temp_in[-1].upper() == 'F': print("Las temperaturas en Fahrenheit tienen que ser mayor de -459.67°F")
                
                if temp_in[-1].upper() == 'K': print("Las temperaturas en Kelvin tienen que ser mayor de 0K")            

#Función principal para agrupar las demás funciones y mostrar el resultado en pantalla.

def main_temp():

    print("Programa para la conversion de temperaturas entre Celsius (c), Fahrenheit (f) e Kelvin (k):\n")

    sigo_con_programa = 'y'

    while sigo_con_programa == 'y':

        respuesta = input("Quieres convertir temperaturas entre Celsius (c), Fahrenheit (f) e Kelvin (k)? (y/n): ").strip().lower()
        
        if keyboard.is_pressed('esc') or respuesta == 'n': print("Salida del programa.") ; sigo_con_programa = 'n'

        elif respuesta == 'y':

            temp_in = obtener_temperatura()

            if temp_in is not None: print(convertir_temperatura(temp_in)) ; sigo_con_programa = 'n'

        else: print("Respuesta no reconocida. Presiona 'y' o 'n' o 'Esc'.")

import keyboard

main_temp()


```

```python
"""Cuenta Palabras"""

#Función para contar las palabras de un texto ingresado por el usuario
#Esta función convierte el texto a minúsculas, reemplaza los signos de puntuación con espacios,
#transforma el texto en una lista de palabras y devuelve un diccionario donde las claves son las palabras encontradas
#y los valores son sus frecuencias.

def comptador_paraules(texto):

    palabras = texto.lower().translate(str.maketrans(string.punctuation + "¡¿", ' ' * (len(string.punctuation) + 2))).split()

    contador = {}

    for word in palabras: contador[word] = contador.get(word, 0) + 1

    return contador

#Función para la introducción del texto por parte del usuario
#Esta función solicita al usuario que ingrese un texto y devuelve el texto ingresado,
#a menos que el usuario presione la tecla Esc, en cuyo caso la función no devuelve nada.

def obtener_texto():

    texto_valido = False 
    
    while not texto_valido:

        texto_in = input("Escribe el texto:\n")
                
        if keyboard.is_pressed('esc'): print("Has presionado Esc.") ; texto_valido = True ; return None

        elif not texto_in: print("No has ingresado ningun texto")

        else: texto_valido = True ; return texto_in

#Función principal para agrupar las demás funciones y mostrar el resultado en pantalla.

def main_cuenta_palabras():

    print("Programa para contar las veces que aparece cada palabra o cada numero en un texto")
    
    sigo_con_programa = 'y'

    while sigo_con_programa == 'y':
        
        respuesta = input("Quieres ingresar un texto para contar las palabras? (y/n): ").strip().lower()

        if keyboard.is_pressed('esc') or respuesta == 'n': print("Salida del programa "); sigo_con_programa = 'n'
        
        elif respuesta == 'y':
            
            texto_in = obtener_texto()

            if texto_in is not None:

                print(f"\nHas ingresado:\n{texto_in}\n\nResultado:")

                for palabra, count in sorted(comptador_paraules(texto_in).items(), key=lambda item: (-item[1], item[0])):
                    
                    print(f"'{palabra}' = {count}")

                sigo_con_programa = 'n'  
        
        else: print("Respuesta no reconocida. Presiona 'y' o 'n'.")


import string
import keyboard
main_cuenta_palabras()
```

```python
"""Diccionario inverso con mensaje de error"""

#Función para la inversión de un diccionario dado
#Esta función intenta invertir las claves y los valores de un diccionario dado.
#Si encuentra valores duplicados que se transformarían en claves, no devuelve nada.
#Si logra invertir todo el diccionario correctamente, lo devuelve.

def diccionario_inverso(diccionario_original):
    
    diccionario_inv = {}
    
    for key, value in diccionario_original.items():
    
        if value in diccionario_inv: return None

        else: diccionario_inv[value] = key
    
    return diccionario_inv

#Función principal para pasar las listas de prueba y llamar a la función de inversión
#Esta función verifica si la función de inversión ha devuelto un diccionario o no y muestra el mensaje correspondiente.

def main_invertir_diccionario_con_error():

    diccionario1 = {'a': 1, 'b': 2, 'c': 3}

    diccionario2 = {'x': 'apple', 'y': 'banana', 'z': 'banana'}

    print("Programa para invertir un diccionario dado con mensaje de error en caso de llaves dobles\n")

    for diccionario in [diccionario1, diccionario2]:

        diccionario_inv = diccionario_inverso(diccionario)

        if diccionario_inv: print(f"Diccionario original:\n{diccionario}\nDiccionario invertido:\n{diccionario_inv}\n\n")
        
        else: print(f"Diccionario original:\n{diccionario}\nError: multiple keys for one value\n")

main_invertir_diccionario_con_error()
```

```python
"""Diccionario inverso con lista de dobles"""

#Función para la inversión de un diccionario dado
#Esta función intenta invertir las claves y los valores de un diccionario dado.
#Si encuentra valores duplicados que deben transformarse en claves, 
#crea una lista donde inserta los valores de las claves duplicadas.
#Siempre devuelve el diccionario invertido.

def diccionario_inverso(diccionario_original):
    
    diccionario_inv = {}
    
    for key, value in diccionario_original.items():
    
        if value in diccionario_inv:
    
            if isinstance(diccionario_inv[value], list): diccionario_inv[value].append(key)
            
            else: diccionario_inv[value] = [diccionario_inv[value], key]
        
        else: diccionario_inv[value] = key
    
    return diccionario_inv

#Función principal para pasar las listas de prueba y llamar a la función de inversión
#Esta función verifica si la función de inversión ha devuelto un diccionario con listas como valores
#o no y muestra el mensaje correspondiente.

def main_invertir_diccionario_con_dobles():

    print("Programa para invertir un diccionario dado con listas para valores dobles\n")
    
    diccionario1 = {'a': 1, 'b': 2, 'c': 3}

    diccionario2 = {'x': 'apple', 'y': 'banana', 'z': 'banana'}

    for diccionario in [diccionario1, diccionario2]:

        diccionario_inv = diccionario_inverso(diccionario)

        print(f"Diccionario original:\n{diccionario}\n")
        
        if any(isinstance(value, list) for value in diccionario_inv.values()): print(f"Error: multiple keys for one value\nDiccionario invertido:\n{diccionario_inv}\n\n")
        
        else: print(f"El diccionario ha sido invertido correctamente\nDiccionario invertido:\n{diccionario_inv}\n\n")
        
main_invertir_diccionario_con_dobles()
```

```python
"""Separa numeros"""

#Esta función procesa cada elemento de una lista dada para separar los elementos numéricos de los demás. 
# En el caso de que el elemento de la lista sea también una lista o una tupla, procesa cada elemento de esta sublista con la misma función. 
# La función devuelve las dos listas con los elementos separados.

def separa_numeros(lista):
   
    lista_de_convertibles = []

    lista_de_no_convertibles = []

    def procesa_elemento(elemento):

        if isinstance(elemento, (tuple, list)): 
            for sub_item in elemento: procesa_elemento(sub_item)

        else:
            try: lista_de_convertibles.append(float(elemento))

            except (ValueError, TypeError): lista_de_no_convertibles.append(elemento)

    
   
    for item in lista: procesa_elemento(item)
    
    return lista_de_convertibles, lista_de_no_convertibles

#Función principal para pasar las listas de prueba a la función de separación 
# y mostrar el resultado en pantalla.

def main_separa_numeros():

    print("Programa para separar los números de los otros valores desde una lista dada\n")
    
    lista_prueba = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']]
    
    lista_convertibles, lista_no_convertibles = separa_numeros(lista_prueba)

    print(f"Elementos convertibles: \n{lista_convertibles}\n")

    print(f"Elementos no convertibles: \n{lista_no_convertibles}\n")

main_separa_numeros()
```

```python
"""Cuenta palabras de un file"""

#Función para contar las palabras de un archivo dado.
#Esta función abre temporalmente un archivo en modo de lectura, 
#transforma todos los caracteres a minúsculas, 
#elimina la puntuación sustituyéndola por espacios, 
#crea una lista de palabras, 
#crea un diccionario donde las claves son las palabras y los valores son sus frecuencias. 
#Luego, crea un diccionario donde las claves son las primeras letras de las palabras encontradas en el archivo, 
#el valor de cada letra es un subdiccionario 
#donde las claves son las palabras que comienzan con esa letra 
#y los valores son las frecuencias del diccionario anterior.

def cuenta_palabras_file():

    import string

    file = 'tu_me_quieres_blanca.txt' 
    
    contador = {}

    try:
        with open(file, 'r', encoding='utf-8') as file:

            palabras = file.read().lower().translate(str.maketrans(string.punctuation + "¡¿", ' ' * (len(string.punctuation) + 2))).split()

            for word in palabras: contador[word] = contador.get(word, 0) + 1

        diccionario_ordenado = {}

        for word in sorted(contador.keys()): 
            
            primeraletra = word[0]
            
            if primeraletra not in diccionario_ordenado: diccionario_ordenado[primeraletra] = {}

            diccionario_ordenado[primeraletra][word] = contador[word]

        for letra, palabras in diccionario_ordenado.items():

            print(f"\nPalabras con '{letra}':")

            for palabra, count in palabras.items(): print(f"{palabra}: {count}")

    except FileNotFoundError: print(f"Error: el archivo '{file}' no fue encontrado.")

cuenta_palabras_file()
```
